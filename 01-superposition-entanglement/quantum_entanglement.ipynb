This notebook creates a quantum circuit with two qubits:
- The first qubit is placed in superposition using the Hadamard gate (H)
- Then it is entangled with the second qubit using the CNOT gate
- Both qubits are measured
- The output is a histogram showing the probability distribution

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from qiskit_aer import Aer # Correct import for Aer
# No need to import execute from qiskit anymore as it's deprecated in favor of backend.run()

# Create a circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2, 2)

# Step 1: Apply Hadamard gate to qubit 0 to create superposition
qc.h(0)

# Step 2: Apply CNOT gate to entangle qubit 0 with qubit 1
qc.cx(0, 1)

# Step 3: Measure both qubits
qc.measure([0, 1], [0, 1])

# Optional: Visualize the circuit
# Ensure pylatexenc is installed before this line
qc.draw('mpl')


```

```python
# Run the simulation with 1000 shots (executions)
backend = Aer.get_backend('qasm_simulator')
# Use the backend.run() method instead of the deprecated execute function
result = backend.run(qc, shots=1000).result()
counts = result.get_counts()

# Display the result as a histogram
plot_histogram(counts)
plt.title("Result: Superposition + Entanglement")
plt.show()
```
